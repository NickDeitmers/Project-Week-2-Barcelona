```python
gender_col = ['year', 'Codi_Districte', 'Nom_Districte', 'Codi_Barri', 'Nom_Barri',
       'gender', 'age', 'Nombre']
gender_age.columns = gender_col
```

```python
gender_age["age"] = gender_age["age"].str.replace(r"\s.*anys", "")
gender_age["age"] = gender_age["age"].str.replace(r"\s.*i m√©s", "")
gender_age["age"] = gender_age["age"].str.replace(r"\s.*any", "")
gender_age["age"] = gender_age["age"].str.replace(r"^0", "1")

gender_age["age"] = pd.to_numeric(gender_age["age"], errors="coerce")
gender_age["Nombre"] = pd.to_numeric(gender_age["Nombre"], errors="coerce")
gender_age["age_product"] = gender_age["age"] * gender_age["Nombre"]

x = pd.DataFrame(gender_age.groupby(["Nom_Districte"])["age_product"].sum())
age = pd.DataFrame(gender_age.groupby(["Nom_Districte"])["Nombre"].sum())
age["avg_age"] = x["age_product"] / age["Nombre"]
age.columns = ["Population", "AVG Age"]
```

```python
age = age.sort_values(by = "Population", ascending = False)
age
```
